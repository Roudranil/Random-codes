```python
import numpy as np
import math
import matplotlib.pyplot as plt 
plt.rcParams['figure.figsize'] = 16, 16
```

```python
def mandplot(n, steps):
    X = np.linspace(start=-2.01, stop=0.48, num=n)
    Y = np.linspace(start=-1.13, stop=1.13, num=n)

    img = np.zeros((n,n))

    for nx in range(n):
        for ny in range(n):
            it = divtest(complex(X[nx], Y[ny]), steps)
            img[ny][nx] = 255-it
    
    return img

def divtest(c:complex, steps):
    i = 1
    z = c
    while i<steps and abs(z) < 2:
        z = z*z + c
        i+=1
    return i

img = mandplot(5000, 50)
plt.imshow(img, cmap='bone', vmin=0, vmax=255)
plt.grid()
plt.show()
```
